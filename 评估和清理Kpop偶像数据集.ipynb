# 简介

数据集提供了1700多位K-Pop（韩流）偶像的信息，包括他们的艺名、全名、韩文名、出生日期、身高、体重、出生地以及之前和其他团体的关联。K-Pop是全球最受欢迎和最具活力的音乐产业之一，这个数据集适合韩流产业或者进行偶像研究的数据分析人员。


变量含义：
- `Stage Name`：偶像被熟知的舞台上名字，即艺名
- `Full Name`：偶像的全名
- `Korean Name`：偶像的韩文名字
- `K Stage Name`：偶像的韩文舞台名字
- `Date of Birth`：偶像的出生日期
- `Group`：偶像所属的韩流团体
- `Debut`：偶像出道的日期
- `Company`：偶像所属娱乐公司
- `Country`：偶像的原籍国家
- `Second Country`：偶像的第二个国家（如果有的话）

# 读取数据

导入分析所需要的数据库，并通过pandas的read_csv函数，将原始数据文件`e_commerce.csv`里的数据内容解析为DataFrame，并赋值给变量`original_data`。

```python
import pandas as pd
```

```python
original_data=pd.read_csv("./kpop_idols.csv")
original_data.head(10)
```

# 评估数据

在这一部分，我将基于上述已经建立好的`original_data`这个DateFrame所包含的数据进行评估。
评估主要从两个维度展开：
- **结构性的整齐度**：数据是否符合“每一列是一个变量，每一行是一个观察值，每一个单元格是一个值”这三个标准。
- **内容性的干净度**：数据中是否存在丢失数据、重复数据以及无效数据。

```python
original_data.sample(10)
```

从抽样的10条数据来看 ，数据符合"每一列是一个变量，每一行是一个观察值，每一个单元格是一个值"的标准。具体来看，每行是一个爱豆的公开个人信息，每列是爱豆的具体信息，因此不构成结构上的问题。

### 评估数据的干净度

在众多数据中优先需要观察的是`Debut`,`Company`,`Group`,`Country`字段完整且一致，这是分析行业趋势的基石。

```python
original_data.info()
```

从输出结果来看，数据总共存在1778条观察值。根据对观察值的重要性要求来看，`Debut`,`Company`存在空缺值。此外，`Date of Birth`的数据类型应该为日期时间而非字符串，应当对数据进行格式转换。

#### 评估缺失数据

在了解`Debut`存在缺失值之后，根据条件提取缺失观察值。

```python
original_data[original_data["Debut"].isnull()]
```

存在146条爱豆个人信息中缺失`Debut`变量值。

从输出结果来看，这些缺失`Debut`的个人数据，同时也缺失`Company`和`Group`变量值。为了检验猜想，我们将增加筛选条件，查看是否存在`Debut`变量值缺失且`Company`与`Group`变量值没有缺失的情况

```python
original_data[original_data["Debut"].isnull() & original_data["Company"].notnull() & original_data["Group"].notnull()]
```

筛选出来的结果为0条，说明缺失`Debut`变量值的数据，同时也不具备有效的`Company`和`Group`值。具体来说，`Debut`表示爱豆个人或者其所属团体的出道日期，而`Company`和`Group`分别指代爱豆所属的娱乐公司和所属偶像团体，都是进行后续偶像行业分析的重要变量。倘若它们同时存在缺失/无效，我们有理由认为该类数据无法提供有效的意义，因此这些数据可以在后期被删除。

`Former Group`也同样存在缺失值的情况，因此需要根据条件提取缺失观察值。

```python
original_data[original_data["Former Group"].isnull()]
```

`Former Group`表示爱豆之前出道的团体，即存在二次出道的情况，但这并不是分析爱豆行业发展趋势的必要变量，且从输出结果来看有些缺失该变量的信息仍然有效，因此保留此变量为空的观察值。

#### 评估重复数据

检查所有列完全重复的行

```python
original_data[original_data.duplicated()]
```

由上述查重步骤可知，所有列中完全重复的行为0条，即不存在单个爱豆个人信息的重复，但考虑到`Stage Name`可能会存在重复的可能性，因此可以检查特定列的重复值。

```python
original_data[original_data.duplicated(subset="Stage Name",keep=False)]
```

由此可见存在520条爱豆个人数据中的`Stage Name`值有重复的情况。但由于韩国人名字存在高度重复率的可能性，即一个相同艺名可能对应的是不同的爱豆实例，因此可以保留该数据。

### 评估不一致数据

不一致数据可能存在于`Company`变量中，我们需要检查是否存在多个值指代同一个公司的情况。

```python
original_data["Company"].value_counts()
```

从变量值来看，猜想未能成立，`Country`值不存在数据不一致的问题。还需要再检查一下`Second Country`中是否存在不同的值指向一个国家的可能性。

```python
original_data["Second Country"].value_counts()
```

根据上述步骤可以知晓，`Second Country`变量值确实存在问题，主要集中在：
- 国家名称不一致，如`USA`和`United States`同时存在
- 组合值格式混乱，分隔符不一致
- 包含国家非国家实体，城市名混入国家列
- 国家和地区归属争议，如`Taiwan`和`HongKong`作为独立值出现，应该同时归属为`China`

为了确保`Country`,`Second Country`两个变量中的值格式一致，需要再查询`Country`中是否存在数据不一致的问题。

```python
original_data["Country"].value_counts()
```

根据输出内容可以发现存在表述上的值和`Second Country`不一致的问题，需要在后续进行以下调整：
- `USA`替换成`Unied States`
- `HongKong`,`Macau`,`Taiwan`替换成`China HongKong`,`China Macau`,`China Taiwan`

#### 评估无效或错误数据

采用`describe`方法，对数据统计信息首先快速了解

```python
original_data.describe()
```

从输出结果来看，`Weight`值存在异常数据，即最大值==170，或许会对后续数值分析产生影响。因此，我们优先筛选出`Weight`数值异常的观察值。

```python
original_data[original_data["Weight"]==170]
```

从输出结果看，`Weight`变量为170的观察值存在`Height`变量有空缺值的情况。根据爱豆行业对身高体重的控制要求来看，体重不可能达到170的情况，因此有可能存在`Weight`值和`Height`值混淆的可能。
需要对数值进行调整。

但此猜想不一定准确，为了验证猜想，我们需要缩小筛选范围即`Weight`>100,并增加筛选条件，查看是否存在`Weight`变量大于100，且`Height`不为空值的值。

```python
original_data[(original_data["Weight"]>100) & (original_data["Height"].notnull())]
```

根据输出结果，可以得知猜想正确，即爱豆体重不可能超过100，甚至连张兵（官方数据）都做不到。故此，极有可能存在`Weight`与`Height`两个变量值混淆的可能性，需要在后续步骤中进行清理。

# 清理数据

根据上述的数据评估得出的结论，我们需要进行的数据清理包括：
- `Date of Birth`变量的数据类型需要转换为日期时间。
- 把同步缺失`Debut`,`Company`,`Group`这三个变量的观察值删除。
- 把`Second Country`中的分割符进行统一，删除空格。
- 把`Second Country`变量值的`USA`替换成`United States`。
- 把`Second Country`变量值的`Seoul`替换成`South Korea`。
- 把`Second Country`变量值中的`Taiwan`和`HongKong`替换为`China Taiwan`和`China HongKong`。
- 把`Country`变量值中的`USA`替换成`United States`。
- 把`Country`变量值中的`Taiwan`和`HongKong`,`Macau`替换为`China Taiwan`和`China HongKong`,`China Macau`。
- 对`Weight`,`Height`中存在异常的数值进行调整。

```python
cleaned_data=original_data.copy()
cleaned_data.head(10)
```

把`Date of Birth`变量的数据类型转换为日期时间

```python
cleaned_data["Date of Birth"]=pd.to_datetime(cleaned_data["Date of Birth"])
cleaned_data["Date of Birth"]
```

把同步缺失`Debut`,`Company`,`Group`这三个变量的观察值删除,并查看删除后该列空缺值的个数和。

```python
cleaned_data.dropna(subset=["Debut"],inplace=True)
```

```python
cleaned_data["Debut"].isnull().sum()
```

把`Second Country`变量值的`USA`替换成`United States`。

把`Second Country`变量值的`Seoul`替换成`South Korea`。

并且检查替换后的两个值的数量。

```python
cleaned_data["Second Country"]=cleaned_data["Second Country"].replace({"USA":"United States","Seoul":"South Korea"})
cleaned_data
```

```python
len(cleaned_data[cleaned_data["Second Country"] =="USA"])
```

```python
len(cleaned_data[cleaned_data["Second Country"]=="Soeul"])
```

把`Second Country`变量值中的`Taiwan`和`HongKong`替换为`China Taiwan`和`China HongKong`。

```python
cleaned_data["Second Country"]=cleaned_data["Second Country"].replace({"Taiwan":"China Taiwan","HongKong":"China HongKong"})
```

```python
len(cleaned_data[cleaned_data["Second Country"]=="Taiwan"])
```

对`Weight`,`Height`中存在异常的数值进行调整,使其数值互换，形成合理的数据。

```python
mask=(cleaned_data["Weight"]==170) & (cleaned_data["Height"].isnull())
affected_rows=cleaned_data[mask]
```

```python
cleaned_data.loc[mask,"Height"]=cleaned_data.loc[mask,"Weight"]
cleaned_data.loc[mask,"Weight"]=pd.NA
```

```python
print(cleaned_data.loc[mask,["Stage Name","Height","Weight"]])
```

```python
cleaned_data["Second Country"]=cleaned_data["Second Country"].str.replace(r"\s*\|\s*","|",regex=True)
```

```python
cleaned_data["Second Country"].value_counts()
```

把`Country`变量值中的`USA`替换成`United States`。

```python
cleaned_data["Country"]=cleaned_data["Country"].replace({"USA":"United States","Taiwan":"China Taiwan","Macau":"China Macau","HongKong":"China HongKong"})
```

```python
print(len(cleaned_data[cleaned_data["Country"]=="USA"]))
print(len(cleaned_data[cleaned_data["Country"]=="Taiwan"]))
print(len(cleaned_data[cleaned_data["Country"]=="Macau"]))
print(len(cleaned_data[cleaned_data["Country"]=="HongKong"]))
```

# 保存清理好的数据

完成数据的清理之后，把干净整齐的数据保存到新的文件中，文件名为`kpop_idols_cleaned.csv`

```python
cleaned_data.to_csv("kpop_idols_cleaned.csv",index=False)
```
